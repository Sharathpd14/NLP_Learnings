### Using Bag of words

```python
import pandas as pd
```

```python
message = pd.read_csv('spam.csv')
```

```python
message = message.iloc[:,:2]
```

```python
message.columns = ["Label","Message"]
```

```python
message
```

```python
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

PS = PorterStemmer()
```

```python
corpus = []

for i in range(0,len(message)):
    review = re.sub('[^a-zA-Z]'," ",message["Message"][i])
    review = review.lower()
    review = review.split()
    review = [PS.stem(word) for word in review if word not in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

#### Create Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500,ngram_range=(1,2))

X = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np

np.set_printoptions(edgeitems=30,linewidth=100000, formatter=dict(float = lambda x : '%.3g' % x))

X  # independent features
```

```python
cv.vocabulary_
```

```python
# Dependent feature or Ouptut feature
Y = pd.get_dummies(message['Label'])
```

```python
Y = Y.astype(int)
```

```python
Y
```

```python
Y = Y.iloc[:,0].values
```

```python
# Train test split

from sklearn.model_selection import train_test_split
X_train,X_Test,Y_Train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=21)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(X_train,Y_Train)
```

```python
Y_predict = spam_detect_model.predict(X_Test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(Y_test,Y_predict)
```

```python
print(classification_report(Y_test,Y_predict))
```

### Using TF-IDF model

### Best Practice ( Prevents data leakage)
##### step 1 : Data Preprocessing and cleaning
##### step 2 : Train Test Split
##### step 3 : Bag of words model (sentences to vectors)
##### step 4 : Train our model

```python
# Train test split

from sklearn.model_selection import train_test_split
X_train,X_Test,Y_Train,Y_test = train_test_split(corpus,Y,test_size=0.20,random_state=21)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(max_features=2500,ngram_range=(1,2))

X_train = tv.fit_transform(X_train).toarray()
X_Test = tv.transform(X_Test).toarray()
```

```python
X_train
```

```python
tv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB

spam_detect_model = MultinomialNB().fit(X_train,Y_Train)
```

```python
## Predictions

y_pred = spam_detect_model.predict(X_Test)
```

```python
accuracy_score(Y_test,y_pred)
```

```python
print(classification_report(Y_test,y_pred))
```
